```python
from sklearn import neighbors
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split, cross_val_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import preprocessingFuncts as pp
import sklearn.metrics as metrics
```

```python
ratingData = pp.readRatingData('ml-100k\\ua.base').merge(pp.readItemData().drop(['title', 'year', 'year_category'], axis='columns'))
xTrain, yTrain = ratingData.drop(
    ['rating'], axis='columns'), ratingData.loc[:, 'rating']
# ratingData = ratingData.pivot_table(
#     index='user_id', columns='item_id', values='rating')

```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import classification_report
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(xTrain,yTrain)

testData = pp.readRatingData('ml-100k\\ua.test').merge(pp.readItemData().drop(['title', 'year', 'year_category'], axis='columns'))
testX, testY = testData.drop(['rating'],axis='columns'), testData.loc[:, 'rating']

predY = model.predict(testX)
print(classification_report(testY, predY))


cm = confusion_matrix(
    testY, predY, labels=[1, 2, 3, 4, 5]
)
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm, display_labels=[1, 2, 3, 4, 5])
disp.plot()
plt.show()
```
